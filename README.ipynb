# AI Ethics Skill-Profiler

Ein Tool zur Analyse von Fairness und Bias in Machine-Learning-Modellen.
Erstellt mit Python, TensorFlow und verantwortungsbewusster KI-Praxis.

## Funktionen
- Lädt ein TensorFlow-Modell
- Berechnet Fairness-Metriken
- Vergleicht Gruppen
- Erstellt einen Bias-Score
- Generiert einen Text-Report

## Ziel
Transparente, faire und verantwortungsvolle KI-Modelle fördern.

## Benutzung
1. Modell trainieren: `python src/train_model.py`
2. Analyse starten: `python src/analyze_fairness.py`

## Ordnerstruktur
- `src/` → Code
- `models/` → gespeicherte Modelle & Testdaten
- `reports/` → Fairness-Reports
